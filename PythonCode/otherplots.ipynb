```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import seaborn
%matplotlib inline
```

```python
trueutil=pd.read_csv('../MABMaxDiff/HB_120items.csv').get_values().mean(0)[2:]
```

```python
values=pd.read_csv('../MABMaxDiff/HB_120items.csv').get_values()[:,2:]
```

```python
filename="newTSthres-120v20k20"
data=pd.read_csv('Results/choice_data-'+filename+'.csv')
choices=data[['Choice1','Choice2', 'Choice3', 'Choice4', 'Choice5']].get_values()-1
```

```python
realmean=np.mean(values,0)
```

```python
order=np.argsort(realmean)[::-1]
```

```python
number=np.bincount(choices.flatten())
```

```python
numbered=number[order]/np.sum(number)
```

```python
plt.bar(range(0,120),numbered)
plt.title('Frequency of Item Shown for $\epsilon$-$\delta$-diffuse TS with thresholding')
plt.ylabel('Frequency shown')
plt.xlabel('Item')
plt.ylim([0,.05])
plt.savefig('edTSthresfreq.pdf')
```

```python
k=10
itemcol=[]
for i in range(1,120+1):
    itemcol+=['Item_'+str(i)]
topk=np.argsort(trueutil)[::-1][:k]
filename="newTSregthres-120v20k10"
bhist=pd.read_csv('Results/bhist-'+filename+'.csv')
itemsin=[]
itemsout=[]
for i in range(100):
    cur=bhist[np.logical_and(bhist['Iter']==(i+1),bhist['RespNum']==500)][itemcol].get_values()[0]
    topkc=np.argsort(cur)[::-1][:k]
    itemsin+=[((set.difference(set(topkc),set(topk))))]
    itemsout+=[((set.difference(set(topk),set(topkc))))]
```

```python
for a,b in zip(itemsin,itemsout):
    print(a,b)
```
